```python
import numpy as ny
import matplotlib.pyplot as plt
from google.colab import files
import pandas as pd
import io
 
data = files.upload()
```

```python
df = pd.read_csv(io.StringIO(data['cost_file.csv'].decode('utf-8')))
```

```python
result_list = []

for i, row in df.iterrows():
  if(df['TYPE_shipment'].iat[i] == 1):
    if(df['X_total_weight'].iat[i] > df['weightslab'].iat[i]):
      x = (row['X_total_weight']/row['weightslab'])-1
      result=(row['forward_fix']) + x*row['forward_addition']
      result_list.append(result)
    else:
      result=(row['X_total_weight'])*(row['forward_fix'])
      result_list.append(result)
  else:
    if(df['X_total_weight'].iat[i] > df['weightslab'].iat[i]):
      x = (row['X_total_weight']/row['weightslab'])-1
      result=((row['fixed_rto']+row['forward_fix'])) + x*(row['additional_rto']+row['forward_addition'])
      result_list.append(result)
    else:
      result=(row['X_total_weight'])*(row['fixed_rto'])
      result_list.append(result)

df['total_cost']=result_list

print(df)
```
